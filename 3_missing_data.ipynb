```python
import numpy as np 
import pandas as pd 
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum() # checking total nan values in every column 
```

```python
df.isna().any() # checking does every column has nan values
```

 Removing Missing Data

```python
df
```

```python
df.dropna() # drop rows with nan values
```

```python
df.dropna(thresh=3) # don't drop rows with at least 3 non-nan values
```

Filling the missing Data

```python
df
```

```python
df.fillna(0) # fill missing values with 0 and if you add inplace = True it will change permanent
```

```python
dict = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=dict) # fill nan values with given dict values according to given dict
```

```python
df.fillna(df.mean()) # fill with mean
```
